```python
import re
import ast
data['Description'] = data['Description'].apply(ast.literal_eval)

# Function to extract values from dictionary
def extract_value(row, key):
    return row.get(key, None)

# List of keys to extract
keys = ['Processor', 'RAM', 'OS', 'SSD', 'Display', 'Warranty']

# Create new columns
for key in keys:
    data[key] = data['Description'].apply(lambda x: extract_value(x, key))

# Define the conditions for removal
conditions = (
    data['Processor'].str.contains('Processor', case=False, na=False) &
    data['RAM'].str.contains('RAM', case=False, na=False) &
    data['OS'].str.contains('Operating System', case=False, na=False) &
    data['Display'].str.contains('Display', case=False, na=False) &
    data['Warranty'].str.contains('Warranty', case=False, na=False)
)

# Filter the DataFrame
filtered = data[conditions]

# Display the cleaned DataFrame
print(filtered)
```

```python
# Function to filter laptops based on criteria
def filter_laptops(processor=None, ram=None, os=None, ssd=None, display=None):
    filtered_df = filtered
    
    if processor:
        filtered_df = filtered_df[filtered_df['Processor'].str.contains(processor, case=False, na=False, regex=False)]
    if ram:
        filtered_df = filtered_df[filtered_df['RAM'].str.contains(ram, case=False, na=False, regex=False)]
    if os:
        filtered_df = filtered_df[filtered_df['OS'].str.contains(os, case=False, na=False, regex=False)]
    if ssd:
        filtered_df = filtered_df[filtered_df['SSD'].str.contains(ssd, case=False, na=False, regex=False)]
    if display:
        filtered_df = filtered_df[filtered_df['Display'].str.contains(display, case=False, na=False, regex=False)]
    
    return filtered_df[['Product Name', 'Description', 'Original Price', 'Discounts', 'Discounted Price', 'Reviews', 'Processor', 'RAM', 'OS', 'SSD', 'Display']]

# Example usage
os_filter = 'Windows 11 Operating System'
processor_filter = 'Intel Core i5 Processor (12th Gen)'
ssd_filter = '512 GB SSD'
ram_filter = '8 GB DDR4 RAM'
display_filter = '39.62 cm (15.6 Inch) Display'

filtered_laptops = filter_laptops(processor=processor_filter, ram=ram_filter, os=os_filter, ssd=ssd_filter, display=display_filter)
print("Filtered Laptops:\n", filtered_laptops)
```

```python
def print_unique_values(filtered, columns):
    for column in columns:
        unique_values = filtered[column].unique()
        print(f"Unique values in '{column}':")
        for value in unique_values:
            print(f"  - {value}")
        print("\n")

# Columns to check for unique values
columns_to_check = ['Processor', 'RAM', 'OS', 'SSD', 'Display', 'Warranty']

# Print unique values for the specified columns
print_unique_values(filtered, columns_to_check)
```

```python
def extract_product_name(product_name):
    # Regular expression to match the product name before the processor
    match = re.match(r'^(.*?)(?:Intel|AMD|Apple)', product_name)
    if match:
        return match.group(1).strip()
    return product_name

# Apply the function to the DataFrame
filtered['Product Name'] = filtered['Product Name'].apply(extract_product_name)

print(filtered)
```

```python
filtered['Brand'] = filtered['Product Name'].str.split(' ').str[:1].str.join(' ')

print(filtered)
```

```python
def clean_price(price):
    """
    Remove the currency symbol and commas from a price string.

    Args:
    price (str): The price string to clean.

    Returns:
    float: The cleaned price as a float.
    """
    cleaned_price = price.replace('₹', '').replace(',', '')
    return float(cleaned_price)
```

```python
filtered['original_price'] = filtered['original_price'].apply(clean_price)
filtered['discounted_price'] = filtered['discounted_price'].apply(clean_price)
```
